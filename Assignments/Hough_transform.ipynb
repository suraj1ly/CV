```python
def padding(image,size):
    size=int(size/2)
    len1,len2=image.shape
    pad=[]
    for i in range(0,(len1+2*size)):
        temp=[]
        for j in range(0,(len2+2*size)):
            if i<size or j<size  or j>=(len2+size) or i>=(len1+size):
                temp.append(0)
            else:
                temp.append(image[i-size][j-size])
        pad.append(temp)
    return pad
```

```python
def remove_padding(image,size):
    len1=len(image)
    len2=len(image[0])
    size=int(size/2)
    for i in range(size):
        del image[0]
        
    for i in range(size):
        del image[len1-2*size]

    for i in range(len(image)):
        for j in range(size):
            del image[i][0]
        
        for j in range(size):
            del image[i][len2-2*size]
    return image
    
```

```python
def convert_xy_space(a,b,theta,r):
    theta1=theta*0.0174533
    x1 = a + r*math.cos(theta1)
    y1 = b + r*math.sin(theta1)
    return round(x1),round(y1)
```

```python
def convert_ab_space(x1,y1,theta,r):
    theta1=theta*0.0174533
    a = x1 - r*math.cos(theta1)

    b = y1 - r*math.sin(theta1)
    return round(a),round(b)
```
