```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
#https://docs.opencv.org/3.1.0/dd/d53/tutorial_py_depthmap.html
img1 = cv2.imread('left.jpg',0)
# imgL=cv2.resize(imgL,(500,500))
#Reading the first image
#Reading the second image
img2 = cv2.imread('right.jpg',0)
# imgR=cv2.resize(imgR,(500,500))
stereo_val = cv2.StereoBM_create(numDisparities=80, blockSize=21)
disparity_value = stereo_val.compute(img1,img2)
#Ploting the depth map image 
plt.imshow(disparity_value,'gray')
plt.show()
```
